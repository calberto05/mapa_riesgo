```python
import pandas as pd
```

```python
df = pd.read_excel('coordenadas_trafico.xlsx', skiprows=1)
df.head()
```

```python
# drop column id
df.drop('Destino', axis=1, inplace=True)
```

```python
df.head()
```

```python
df_rename = df.rename(columns={'ini_lat': 'x1', 'ini_lon': 'y1', 'dest_lat': 'x2', 'dest_lon': 'y2'})
```

```python
df_rename.head()
```

```python
# quitar , a todos los elementos de x1, y1, x2, y2 si no son nan
df_rename['x1'] = df_rename['x1'].apply(lambda x: x.replace(',', '') if not type(x) == float else x)
```

```python
# utrn x1, y1, x2, y2 to float
df_rename['x1'] = df_rename['x1'].astype(float)
df_rename['y1'] = df_rename['y1'].astype(float)
df_rename['x2'] = df_rename['x2'].astype(float)
df_rename['y2'] = df_rename['y2'].astype(float)
```

```python
df_rename.head()
```

```python
# save to csv
df_rename.to_csv('coordenadas_trafico.csv', index=False)
```
